```python
engine = BacktestEngine(db_name='./DB/base.db')
```

```python
class myStretegy(Strategy):
    def __init__(self):
        super().__init__()
        
    def compute_target(self, universe_list):
        target_weight = {}
        for ticker in universe_list:
            target_weight[ticker] = 1
            
        target_weight = self.normalize(target_weight)
        
        return target_weight
```

```python
mys = myStretegy()
engine.run_backtest(strategy=mys, sdate='2017-01-01', edate='2020-01-01')
```

```python
def report():
    dates = engine.asset_df.index

    asset = engine.asset_df.sum(axis=1)
    bench = engine.cache['index']['^SP500TR'].closeadj.loc[dates]
    bench = bench/bench.iloc[0]
    
    df = pd.concat([pd.Series(stat(asset, bench)),pd.Series(stat(bench, bench))],axis=1)
    df.columns = ['Strategy','Benchmark']

    fig, axs = plt.subplots(1, 2, gridspec_kw={'width_ratios': [2, 1]},figsize=(12,5))
    
    ax1 = axs[0]
    ax1.plot(bench)
    ax1.plot(asset)
    ax1.legend(['Benchmark', 'Strategy'])
    
    ax2 = axs[1]
    ax2.axis('off')
    table = ax2.table(cellText=df.values, rowLabels = df.index, colLabels=df.columns, bbox=[0,0,1,1])
    table.auto_set_font_size(False)
    table.set_fontsize(10)



    fig.autofmt_xdate()
    
report()
```

```python
def max_underwater(asset):
    ser = (asset - asset.cummax())

    underwater_periods = []
    days = 0
    sdate = ser.index[0]
    for date in ser.index:
        item = ser.loc[date]
        if item < 0:
            days += 1
        else:
            underwater_periods = underwater_periods+[days]
            days=0
            sdate = date
            
    return np.max(underwater_periods)

def stat(asset, bench):
    rets = asset.apply(np.log).diff().dropna()
    excess_rets = asset.apply(np.log).diff().dropna() - bench.apply(np.log).diff().dropna()
    cagr = 252*np.mean(rets)
    vol = np.sqrt(252)*np.std(rets)
    sharpe = np.sqrt(252)*np.mean(rets)/np.std(rets)
    IR = np.sqrt(252)*np.mean(excess_rets)/np.std(excess_rets)
    mdd = (asset - asset.cummax()).min()
    mup = max_underwater(asset)
    
    stat = {'CAGR':'{:.3f}'.format(cagr), 
            'Vol':'{:.3f}'.format(vol), 
            'SR':'{:.3f}'.format(sharpe), 
            'IR':'{:.3f}'.format(IR), 
            'MDD':'{:.3f}'.format(mdd), 
            'MUP':'{} days'.format(mup)}
    
    return stat
```
